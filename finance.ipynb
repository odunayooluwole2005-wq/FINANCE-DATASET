```python
# Cell 1 — Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
# This formatter will make y-axis show real numbers like $25,000,000
# instead of scientific notation like 2.5e7
money_fmt = mticker.FuncFormatter(lambda x, _: f"${x:,.0f}")
print('Ready!')
```

```python
# Cell 2 — Load the dataset
df = pd.read_excel(r"C:\Users\HomePC\Documents\finance_transactions_dataset.xlsx")
# Quick check: how many rows and columns?
print(df.shape) # e.g. (1000, 14)
print(df.columns.tolist())
```

```python
# Cell 4a — Remove rows where essential columns are empty
df.dropna(subset=['Amount', 'Net_Amount', 'Transaction_Date',
'Approval_Status'],
inplace=True)
print('Rows remaining:', len(df))
```

```python
# Cell 4b — Fill gaps in text columns
df['Company'] = df['Company'].fillna('Unknown')
df['Department'] = df['Department'].fillna('Unknown')
df['Analyst'] = df['Analyst'].fillna('Unknown')
df['Category'] = df['Category'].fillna('N/A')
```

```python
# Cell 4c — Turn text dates into proper dates, then extract Year and Month
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'],errors='coerce')
df.dropna(subset=['Transaction_Date'], inplace=True)
df['Year'] = df['Transaction_Date'].dt.year
df['Month'] = df['Transaction_Date'].dt.to_period('M').astype(str)
print('Sample dates:')
print(df[['Transaction_Date', 'Year', 'Month']].head(3))
```

```python
df.drop_duplicates(inplace=True)
print('Final row count:', len(df))
```

```python
# Cell 5a — Bar chart
df.groupby('Transaction_Type')['Amount'].sum().sort_values().plot(kind='bar',
color='steelblue')
plt.title('Total Amount by Transaction Type')
plt.ylabel('Total Amount ($)')
plt.xticks(rotation=15)
plt.gca().yaxis.set_major_formatter(money_fmt) # show real numbers, not
2.5e7
plt.tight_layout()
plt.show()
```

```python
# Cell 5b — Grouped bar chart
df.groupby('Sector')[['Net_Amount', 'Tax_Amount']].sum().plot(kind='bar',color=['steelblue', 'coral'])
plt.title('Net Amount vs Tax Amount by Sector')
plt.ylabel('Amount ($)')
plt.xticks(rotation=20)
plt.gca().yaxis.set_major_formatter(money_fmt) # show real numbers, not 2.5e7

plt.tight_layout()
plt.show()
```

```python
# Cell 5c — Line chart
df.groupby('Month')['Amount'].sum().plot(kind='line', marker='o',
color='steelblue')
plt.title('Monthly Transaction Volume')
plt.ylabel('Total Amount ($)')
plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(money_fmt) # show real numbers, not 2.5e7
plt.tight_layout()
plt.show()
```

```python
# Cell 5d — Pie chart
df['Approval_Status'].value_counts().plot(kind='pie', autopct='%1.1f%%',
startangle=140)
plt.title('Approval Status Distribution')
plt.ylabel('') # hides the default y-axis label on pie charts
plt.tight_layout()
plt.show()
```

```python
# Cell 5e — Horizontal bar chart
df.groupby('Analyst')['Amount'].sum().nlargest(5).plot(kind='barh',
color='mediumseagreen')
plt.title('Top 5 Analysts by Total Amount')
plt.xlabel('Total Amount ($)')
plt.gca().xaxis.set_major_formatter(money_fmt) # x-axis for horizontal bars
plt.tight_layout()
plt.show()
```

```python
# Cell 6 — Summary
print('=' * 46)
print(' FINANCE DATASET SUMMARY')
print('=' * 46)
print(f"Transactions: {len(df):,}")
print(f"Total Amount: ${df['Amount'].sum():,.0f}")
print(f"Total Tax Paid: ${df['Tax_Amount'].sum():,.0f}")
print(f"Total Net Amount: ${df['Net_Amount'].sum():,.0f}")
print(f"Avg Tax Rate: {df['Tax_Rate'].mean()*100:.1f}%")
print(f"Approval Rate:{(df['Approval_Status']=='Approved').mean()*100:.1f}%")
print(f"Top Sector: {df.groupby('Sector')['Amount'].sum().idxmax()}")
print(f"Top Analyst: {df.groupby('Analyst')
['Amount'].sum().idxmax()}")
print('=' * 46)
```

```python
# Group by Currency and sum Amount
currency_totals = df.groupby('Currency')['Amount'].sum()

# Sort values (optional but recommended)
currency_totals = currency_totals.sort_values()

# Plot bar chart
currency_totals.plot(kind='bar')

# Add labels and title
plt.xlabel('Currency')
plt.ylabel('Total Amount')
plt.title('Total Amount by Currency')

# Show chart
plt.show()
```

```python
# Filter only approved transactions
approved = df[df['Approval_Status'] == 'Approved']

# Group by Month and sum Net_Amount
monthly_net = approved.groupby('Month')['Net_Amount'].sum()

# Plot line chart
monthly_net.plot(kind='line', marker='o')

plt.xlabel('Month')
plt.ylabel('Total Net Amount')
plt.title('Monthly Net Amount (Approved Transactions)')

plt.show()
```

```python
# Group by Department and calculate average Amount
dept_avg = df.groupby('Department')['Amount'].mean().sort_values()

# Plot horizontal bar chart
dept_avg.plot(kind='barh')

plt.xlabel('Average Amount')
plt.ylabel('Department')
plt.title('Average Amount per Department')

plt.show()
```

```python
# Count transactions by type
type_count = df['Transaction_Type'].value_counts()

# Plot pie chart
type_count.plot(kind='pie', autopct='%1.1f%%')

plt.title('Transaction Type Distribution')
plt.ylabel('')  # removes default label

plt.show()
```

```python
type_count = df['Transaction_Type'].value_counts()

print(type_count)

most_common = df['Transaction_Type'].value_counts().idxmax()

print(most_common)
```

```python

# Filter only Expense transactions
expenses = df[df['Transaction_Type'] == 'Expense']

# Group by Category and sum Amount
category_total = expenses.groupby('Category')['Amount'].sum().sort_values()

# Plot bar chart
category_total.plot(kind='bar')

plt.xlabel('Category')
plt.ylabel('Total Amount')
plt.title('Total Expense Amount by Category')

plt.show()
```
